```python
%matplotlib inline

from numpy import *
import matplotlib.pyplot as plt

def sigmoid(inX):
    return 1.0/(1+exp(-inX))

def sigmoid2(inX):
    return 1.0/(1+exp(-(3*inX)))

def plotSigmoid(inX):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    x = arange(-inX, inX, 0.005)
    y = sigmoid(x)
    ax.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('Sigmoid(y)')
    plt.show()
    
def plotSigmoid2(inX):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    x = arange(-inX, inX, 0.005)
    y = sigmoid2(x)
    ax.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('Sigmoid(y)')
    plt.show()
    
plotSigmoid(5)
plotSigmoid2(5)
```
